```python
test=pd.read_csv("/kaggle/input/hy-data/test.csv")
test
```

```python
alldata=pd.concat([train,test])
alldata
```

```python
alldata2=alldata[test.columns]
alldata2
```

```python
alldata2=alldata2.drop(columns=["ID"])
alldata2
```

```python
alldata2["fe_zn"]=(alldata2["FE"]+1)/(alldata2["ZN"]+2)
alldata2
```

```python
c=pd.get_dummies(alldata2,columns=["COMPONENT_ARBITRARY"])
c
```

```python
a= alldata2.groupby("COMPONENT_ARBITRARY")["ANONYMOUS_1"].agg(["sum","mean","std","median","max","count"])
b= alldata2.groupby("YEAR")["ANONYMOUS_2"].agg(["sum","mean","std","median","max","count"])
c1=c.groupby("YEAR")["COMPONENT_ARBITRARY_COMPONENT1","COMPONENT_ARBITRARY_COMPONENT2","COMPONENT_ARBITRARY_COMPONENT3","COMPONENT_ARBITRARY_COMPONENT4"].agg(["sum","mean","std","count"])
d= alldata2.groupby("COMPONENT_ARBITRARY")["ANONYMOUS_2"].agg(["sum","mean","std","median","max","count"])

alldata2=pd.merge(alldata2,a,on="COMPONENT_ARBITRARY",how="left")
alldata2=pd.merge(alldata2,b,on="YEAR",how="left")
alldata2=pd.merge(alldata2,c1,on="YEAR",how="left")
alldata2=pd.merge(alldata2,d,on="COMPONENT_ARBITRARY",how="left")
alldata2
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
alldata2["COMPONENT_ARBITRARY"]=le.fit_transform(alldata2["COMPONENT_ARBITRARY"])
```

```python
alldata2
```

```python
train2=alldata2[:len(train)]
test2=alldata2[len(train):]
```

```python
#n_splits=10

from sklearn.model_selection import StratifiedKFold
from catboost import CatBoostClassifier
cbc=CatBoostClassifier(verbose=100)
skf=StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
result=0
best_score=0

for train_index,valid_index in skf.split(train2,train["Y_LABEL"]):
    x_train=train2.iloc[train_index]
    x_valid=train2.iloc[valid_index]
    y_train=train["Y_LABEL"].iloc[train_index]
    y_valid=train["Y_LABEL"].iloc[valid_index]
    cbc.fit(x_train,y_train,eval_set=(x_valid,y_valid),early_stopping_rounds=20,cat_features=[0,2])
    result+=cbc.predict_proba(test2)/10
    best_score+=cbc.best_score_["validation"]["Logloss"]/10
```

```python
#변수중요도
pd.Series(cbc.feature_importances_,index=train2.columns).sort_values(ascending=False)
```

```python
sub=pd.read_csv("/kaggle/input/hy-data/sample_submission.csv")
sub
```

```python
sub["Y_LABEL"] = (result[:,1]>0.16).astype(int)
sub.to_csv("jshy.csv",index=False)
```
